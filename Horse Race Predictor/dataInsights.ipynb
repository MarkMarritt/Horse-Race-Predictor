```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv("MarksClean2.csv")
```

```python
import scipy.stats as st
df =df.drop(df[df["Position"] > 30].index)
pos = np.array(df["Position"])
negPos = pos*(-1)
allPos = np.concatenate((pos, negPos), axis = 0)

mu, sigma = np.mean(allPos), np.std(allPos)
print(mu, sigma)
x = np.linspace(-30, 30, 100)
plt.plot(x, st.norm.pdf(x, mu, sigma), color = "red")
maxY = np.count_nonzero(allPos == st.mode(allPos))
plt.hist(allPos, range = (-30, 30), bins= 60, weights=[1/(maxY*18)]*len(allPos))
plt.show()
```
